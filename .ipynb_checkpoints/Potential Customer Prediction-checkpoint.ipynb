```python
#Creating function to plot categorical variable
def pie_plot(df):
    category = []
    
    for col in df.columns:
        if df[col].dtype.name =='category':
            category.append(col)
    plt.figure(figsize=(20,15))
        
    ncol =3
    nrows = (len(category)/ncol) + 1
    
    for col, n in zip(category, np.arange(len(category))):
        ax = plt.subplot(nrows, ncol, n+1)
        df[col].value_counts().plot(kind='pie', autopct='%.2f%%',ax=ax)
    plt.show()
```

```python
#Calculating the number of outlier in each feature of the dataset
des = data_merged.describe()
for col in des.columns:
    q1 = des.loc['25%',col]
    q3 = des.loc['75%',col]
    iqr = q3-q1
    upper_fence = q3 + 1.5*iqr
    lower_fence = q1 -1.5*iqr
    print("Number of outlier in", col ,"feature is", 
          len(data_merged[col][(data_merged[col]>upper_fence) | (data_merged[col]<lower_fence)]))
```

```python
#Imputing outliers with mean value
des1 = data_merged.describe()
columns=['HighestSpend','MonthlyAverageSpend','Mortgage']
for col in columns:
    q1 = des1.loc['25%',col]
    q3 = des1.loc['75%',col]
    iqr = q3-q1
    upper_fence = q3 + 1.5*iqr
    lower_fence = q1 -1.5*iqr
    df_include = data_merged.loc[(data_merged[col] >= lower_fence) & (data_merged[col] <= upper_fence)]
    mean = df_include[col].mean()
    data_merged[col]=np.where(data_merged[col]>upper_fence,mean,data_merged[col])
    data_merged[col]=np.where(data_merged[col] < lower_fence, mean,data_merged[col])
    print("Number of outlier after imputing outlier with mean in", col ,"feature is", 
          len(data_merged[col][(data_merged[col]>upper_fence) | (data_merged[col]<lower_fence)]))
```

# Hypothesis Testing for continuous variable

Ho(Null Hypothesis):There is no significant difference in independent feature with different category of Target variable

H1(Alternate Hypothesis):There is significant difference in independent feature with different category of Target variable

Here we will use independent t test

```python
#Defining the continuous columns
col = data_merged.select_dtypes(['int64','float64']).columns
```

```python
from scipy.stats import ttest_ind
```

```python
#Independent t test
for c in col:
    x = np.array(data_merged[c][data_merged['LoanOnCard']==0.0])
    y = np.array(data_merged[c][data_merged['LoanOnCard']==1.0])
    t, p_value = ttest_ind(x,y,axis=0)
    print('p_value:',p_value)
    if p_value < 0.05:          #assuming significant level at 0.05
        print('There is a significant difference in',c, 'feature with different category of LoanOnCard')
    else:
        print('There is no significant difference in',c, 'feature with different category of LoanOnCard')
```

```python
#Dropping Age, CustomerSince and Zipcode column
data_merged = data_merged.drop(['Age','CustomerSince','ZipCode'],axis=1)
```

# Hypothesis Testing for Categorical Variable

Ho: There is no significant difference in independent feature for different category of target variable(Loan on card)

H1: There is significant difference in independent feature for different category of target variable(Loan on card)

Here we will us Chi Square test

```python
from scipy.stats import chi2_contingency
```

```python
#Defining the columns
col = data_merged.select_dtypes('category').columns
```

```python
#Chi Square Test
for c in col:
    crs_tab = pd.crosstab(data_merged[c],data_merged['LoanOnCard'])
    chi,p_value,dof,expected=chi2_contingency(crs_tab)
    print('P value:', p_value)
    if p_value < 0.05:
        print('There is a significant difference in',c, 'feature for different category of target variable(Loan on card)')
    else:
        print('There is no significant difference in',c, 'feature for different category of target variable(Loan on card)')
```

# Data Preparation and Modeling

```python
#One hot Encoding for Hiddenscore and Level
data_merged = pd.get_dummies(data_merged, columns=['HiddenScore','Level'])
```

```python
#Splitting independent and target variable
x = data_merged.drop('LoanOnCard',axis=1)
y = data_merged['LoanOnCard']
```

```python
#Splitting data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
#Scaling the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
```

```python
#Balancing the dataset
import imblearn
from imblearn.over_sampling import SMOTE
# Resampling the minority class
sm = SMOTE(sampling_strategy='minority', random_state=42)
# Fit the model to generate the data.
oversampled_x, oversampled_y = sm.fit_resample(x_train_scaled, y_train)
oversampled = pd.concat([pd.DataFrame(oversampled_y), pd.DataFrame(oversampled_x)], axis=1)
```

# Comparing all the models 

```python
def get_metrics_score(model):
    
    score_list = []
    
    pred_train = model.predict(oversampled_x)
    pred_test = model.predict(sc.transform(x_test))
    
    train_acc = round(model.score(oversampled_x,oversampled_y)*100,2)
    test_acc = round(model.score(sc.transform(x_test),y_test)*100,2)
    
    train_recall = round(metrics.recall_score(oversampled_y,pred_train)*100,2)
    test_recall = round(metrics.recall_score(y_test,pred_test)*100,2)
    
    train_precision = round(metrics.precision_score(oversampled_y,pred_train)*100,2)
    test_precision = round(metrics.precision_score(y_test,pred_test)*100,2)
    
    train_f1 = round(metrics.f1_score(oversampled_y,pred_train)*100,2)
    test_f1 = round(metrics.f1_score(y_test,pred_test)*100,2)
    
    score_list.append([train_acc,test_acc,train_recall,test_recall,train_precision,test_precision,train_f1,test_f1])
    
    return score_list
```

```python
from sklearn import metrics
#Defining the odels
models = [log_model,log_model_tuned,svm_model,svm_model_tuned,knn_model,knn_tuned]
#Creating empty list to add train and test results
acc_train = []
acc_test = []
recall_train = []
recall_test = []
precision_train = []
precision_test = []
f1_train = []
f1_test = []
#looping through the models
for mod in models:
    j = get_metrics_score(mod)
    acc_train.append((j[0][0]))
    acc_test.append((j[0][1]))
    recall_train.append((j[0][2]))
    recall_test.append((j[0][3]))
    precision_train.append((j[0][4]))
    precision_test.append((j[0][5]))
    f1_train.append((j[0][6]))
    f1_test.append((j[0][7]))    
```

```python
#Creating Dataframe
comparison_frame = pd.DataFrame({'Model':['log_model','log_model_tuned','svm_model','svm_model_tuned','knn_model','knn_tuned'],
                                          'Train_Accuracy': acc_train,'Test_Accuracy': acc_test,
                                          'Train_Recall':recall_train,'Test_Recall':recall_test,
                                          'Train_Precision':precision_train,'Test_Precision':precision_test,
                                          'Train_F1':f1_train,
                                          'Test_F1':f1_test  }) 

#Sorting models in decreasing order of test accuracy
comparison_frame.sort_values(by='Test_Accuracy',ascending=False)
```
